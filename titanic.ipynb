```python
# importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the data
df = pd.read_csv('titanic.csv')
```

```python
rows,columns = df.shape
print(f"Number of Rows: {rows} and Number of Columns: {columns}")
```

```python
# explore the column Names
print(df.columns)
```

```python
# Renaming the columns as they have Capitalized 1st alphabet
df.rename(columns={'Pclass':'passenger_class','SibSp':'sibling_spouse','Parch':'parent_child','PassengerId':'passengerid','Survived':'survived','Name':'name','Sex':'gender','Age':'age','Ticket':'ticket','Fare':'fare','Cabin':'cabin','Embarked':'embarked'},inplace=True)
```

```python
# First 15 rows of the dataset
df.head(15)
```

```python
# Last 20 rows of the dataset
df.tail(20)
```

### Exploratory Data Analysis

```python
# Missing Values
df.isnull().sum() #cabin column has 687 missing values so we will drop it
```

```python
# Dropping the columns that are not required and have majority of missing values
df.drop('cabin',axis=1,inplace=True)
```

```python
# What is the Average age of the passengers 
avg_age = df['age'].mean()
print("Average age of the passengers is: ", avg_age)
```

```python
# What is the Name and age of the Oldest Passenger?
oldest_passenger = df.iloc[df['age'].idxmax()]

name = oldest_passenger['name']
age = oldest_passenger['age']
print(f"The oldest passenger is {name} and he is {age} years old.")
```

```python
# Grouping the Dataset by Passenger's Class
df.groupby('passenger_class').count()
```

```python
# Grouping the Dataset by Survived Passenger
df.groupby('survived').count()
```

```python
# Count of the Numbers of Survivors
Survived_count = len(df[df['survived']==1])
print(f"The number of Survived Passengers are {Survived_count}")
```

```python
# The Number of Male and female Survivors 
male = len(df[(df['survived']==1) & (df['gender']=='male')])
female = len(df[(df['survived']==1) & (df['gender']=='female')])
print(f"The Number of Male Survivors are {male} and number of Female Survivors are {female}")
```

```python
# The Name and age of Youngest Survivor

youngest_survivor = df.iloc[df['age'].idxmin()]
name = youngest_survivor['name']
age = youngest_survivor['age']
print(f"The Name of the youngest Survivor is {name} and he is {age} years old")
```

```python
# what is the 2nd Class passenger's average Age
second_class_pass = df[df['passenger_class']==2]
average_age = second_class_pass['age'].mean()

print(f"The Average age of 2nd class Passenger is {average_age}")
```

```python
# The Number of survivors in 3rd Class
third_class = df[df['passenger_class']==3]
count_survivors = len(third_class[third_class['survived']==1])

print(f"The number of Survivors from 3rd class are {count_survivors}")
```

```python
# The Average age of Passengers Who died in the 1st Class
first_class = df[(df['passenger_class'] == 1) & (df['survived']==0)]
avg_age_died = first_class['age'].mean()

print(f"The number of Passengers from 1st class that didn't Survive are {avg_age_died}")
```

```python
# How many male and female passengers survived in each class?
grouped = df.groupby(['survived','passenger_class','gender'])['passenger_class'].count().reset_index(name='count')
grouped
```

### Visualization

```python
# Creating Pie chart for the distribution of Passengers Class

passenger_class = df.groupby(['passenger_class'])

sizes = [len(passenger_class.get_group(1)),len(passenger_class.get_group(2)),len(passenger_class.get_group(3))]
labels = ["First Class","Second Class","Third Class"]


plt.pie(sizes,labels=labels,autopct= "%1.1f%%")
plt.show()
```

### Conclusion :

In conclusion, the  exploratory data analysis of the Titanic dataset has revealed several interesting insights. We found that passenger class, gender , and age were all important factors in determining survival rates. 

The average age of a passenger was 30 years, while the oldest passenger was 80 years old.The majority of passengers were in third class, with a total of 491 passengers. Out of the 891 passengers, 342 survived.The average age of a second class passenger was 30 years old. In third class, there were 119 survivors. The average age of the survivors who died in first class was 44 years old. Passengers in first class had a higher survival rate than those in second or third class. Female passengers (233) also had a higher survival rate than male passengers(109).
